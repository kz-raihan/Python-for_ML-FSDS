### 1. Arithmetic OPerators 
- ( + , - , *, /, //, % , **)

```python
# Adddition (+)
num1 = 50
num2 = 30

ans = num1 + num2

# subtraction (-)
sub_ans = num1 - num2

print("Addition: ", ans)

print("Subtraction: ", sub_ans)
```

```python
# multiplication (*)
num1 = 7
num2 = 2

mul_ans = num1 * num2

# divison(/)
div_ans = (num1 / num2)


# floor division(//) [below value]

floor_div_ans = (num1 // num2)

print("Multiplicatin: ", mul_ans)

print("Division: ", div_ans)

print("Floor Division: ", floor_div_ans)
```

```python
# modules (%) [Return the remainder of the result]

num1 = 11
num2 = 2

mod_ans = num1 % num2


# power

pow_ans = num1 ** num2


print("Modules : ", mod_ans)
print("Power: ", pow_ans)
```

### 2. Assignment Operator
(=, +=, -=, *= , %=)

```python
# Assignment operator
num = 10 
print(num)

num += 2     # num = num + 2  => 12
print(num)

num -= 3      # num = num - 2 => 12 - 3 => 9
print(num)

num -= 2      # num = num - 2 => 9 - 2 => 7
print(num)

num *= 3     # num = num * 3 => 7 * 3 => 21
print(num)

num %= 4     # num = num % 4 => 21 % 4 =>  
print(num)
```

### 3. Relational Operators
- (<, <=, >, >=, ==, !=)

```python
# Relational Operators (- Return the boolean values)
num1 = 10
num2 = 20

# less than (<)
print(num1 < num2)

# less than or equal
num3 = 20
print(num2 <= num3)

# Greater than 
print(num1 > num2)

# Greater than or equal
print(num3 >= num2)

# Equal 
print(num2 == num1)

# Not Equal

print(num2 != num3)
```

### 4. Logical Operator
- (and, or, not)

```python
# logical operator (- use for compare two or more condition)

num1 = 10
num2 = 20

# and
print("And(Both True): ", num1 and num2)
print("And: ", 0 and 1)
print("And: ", 1 and 1)


# or
print("or: ", num1 or num2)
print("Or (at least only True): ", 0 or 1)
print("Or (at least only True): ", 0 or 0)

# not
print("NOt: ", not num1)
print("NOt: ", not 0)
print("NOt: ", not 1)
```

### 5. Bitwise Operator
- (&, | , ^ , ~, >> , <<)

```python
# bitwise and
print(2 & 3)

# bitwise or
print(2 | 4)

# bitwise not
print(~4)

# right shift
print(2 >> 4)

# left shift

print(2 << 4)
```

### 6. Membership operator

- in 
- not in

```python
# membership operator

str1 = "My name is KzRaihan"

# in membership operator

print('Kz' in str1)


# not in membership operator
print("Ra" not in str1)
```
