### 1. Comments in Python

- Use for code readability and python interpreter ignore this line (comment)

- Two Types
    - 1. single line comment (#)
    - 2. Multiple line comment ("""  """)

```python
# 1. single line comment 
# This is single line comment

print("single line comment ")

# interpreter ignore comment
```

```python
# Multiple line comment (mainly use docustring)

"""  
The below code perform
- Addition of two number
- using only print()

"""

print("The Example of Multiple line comments")
print(10 + 20)
```

### 2. Data type in Python

- int

- float

- string

- complex

- list

- tuple

- sets

- dictionary

```python
# integer (int) -> Non-decimal numbers

print(2)

print(4)

type(2)
```

```python
# floating point (float) -> Decimal numbers

print(3.14)

print(99.99)

print(type(99.99))
```

```python
# string (str) -> text type data

print("Hello")

print(type('Hello'))
```

```python
# complex -> combine of real and imaginary numbers
print(2 + 3j)

type(2+3j)
```

```python
# boolen
print(True)

type(True)

# case-sensative 
# for true can't work
```

```python
# list (collection of different data)

print([1, 2, 3.14])

type([1, 2, 'hello'])
```

```python
# tuple (collection of data)

print((1,3,5))
type((1,3,5))
```

```python
# dictionary data type

print({'key': 'value', "name": "Kz"})

print(type({'key': 'value', "name": "Kz"}))
```
