```python
### Syntax of print() function

"""
- print(data, sep = ' ', end = '\n')

- data => int, float, etc type of data
    - eg: 6, 3.14

- sep => Data to Data separation format
    - here seperation = space [by default]

- end => when statement is end then by default need one new line.

    - '\n' => new line

"""
```

```python
# change the print() function parameters

print(10)

print(10, 20 , 3.14, True)

# here sep = ' ' (space)
#      end = '\n'
```

```python
# change the sep parameter values

# sep = '-'

print(10, 20 , 3.14, True,sep='-')
```

```python
# change the '\n' (new line) parameters

print("Kz")
print("Raihan")

# change end to single line (end = ' ')
print("Kz", end = " ")
print("Raihan")
```

```python
## Apply all 
# data => Python-KzRaihan-True-90.5 (using two print)

print('Python','KzRaihan',sep='-', end='-')

print(True, 90.5, sep="-", end=' ')
```
