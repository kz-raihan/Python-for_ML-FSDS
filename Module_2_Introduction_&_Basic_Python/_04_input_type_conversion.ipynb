### 1. introduction of intput()

- In real world there are two senerio or application
    - 1. Static application
        - No need to take input to perform work
        - eg: calender, clock etc

    - 2. Dynamic application
        - Need input to input to perform task
        - eg: Youtube, Facebook and so on.

- so, Take a input from user use input()

```python
# example of intput()

name = input()

print(name)
print(type(name))


# NOTE:
#  - using input() by default take string type
```

```python
# another example of input() with message

name = input("Enter your name: ")

print("Name: ", name)
```

### 2. Type Conversion or Type-casting.

- Type Conversion:
    - To convert one data type to other data type

- syntax:
    - targer_type(data)
    - eg: int(3.14)


- There are two way to Type-casting
    - 1. implicit type conversion
        - Do internally by python interpreter

    - 2. Explicit type conversion
        - Do the programmer

```python
# Example of 1. implicit type-casting

num = 10

ans = num + 5.5

print(ans)
print(type(ans)) # this ans type define internally by python interpreter.  

# how?
# smaller data type convert to upper data type 
```

```python
# Example - 2. Explicit Data type
# syntax: targer_type(data)

num = 10 

# ans = num + '4'  <--- for this ouput show Error

ans = num + int('4')

print(ans)
print(type(ans))
```

```python
# another way

var = "Hi"

ans = var + str(10)

print(ans)
```

```python
# Example of type-casting for intput()

var = int(input("Enter a number: "))

print(var, type(var))
```

```python
# some case

var_1 = input()

var_2 = input()

result = var_1 + var_2  # if at least input is str then it's string concatenation

print(var_1, var_2)
print(result, type(result))
```

```python
# some case

var_1 = input()

var_2 = input()

print(var_1, var_2)

result = var_1 - var_2  # for every input it's generated error

print(result, type(result))
```

```python
# To solve the problem use type-casting

num1 = int(input())
num2 = int(input())

result = num1 - num2

print(num1, num2)
print(result)
```

### - simple Task : Create a simple calculator for two number form user that perform addition and substraction
