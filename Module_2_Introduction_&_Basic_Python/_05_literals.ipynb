## Topic: Literals in Python

### OUTCOMES

- 1. Intro

- 2. Type of Literals

- literals Intro:
    - Value inside the Variable is called literals.

    - eg: num = 10 (here, 10 is literals)

- There are many types of literals

    - Decimal literals (10,20)

    - float literals (10.55)

    - string literals 

    - Binary literals (0b1001)

    - octal literals (0o310)

    - hex-decimal (0x12c)

### Decimal literals (10,20)

```python
num1 = 10

age = 24

val = 100


print(num1, age, val)

print(type(num1), type(age), type(val))
```

### float literals (10.55)

```python
float_1 = 10.5

price = 99.9

pi_val = 3.14

var_float = 1.5e2 # 1.5 * 10^2

neg_var = 1.5e-3   # 1.5 * e^-3

print(float_1, price, pi_val)
print(type(float_1), type(price), type(pi_val))

print(var_float, neg_var)
print(type(var_float), type(neg_var))
```

### string literals 

```python
str1 = 'Hello'

str2 = "My Name is KzRaihn"

str3 = """
     - I am from Bangladesh
     - I am a Learner
     - Currently I am learning Python programming language.
"""

print(str1, type(str1))

print(str2, type(str2))

print(str3, type(str3))
```

```python
# Other string literals
# unicode to represent text in any human language and also symbols like emojis.

# 8-digit unicode means emojis
# link : https://unicode.org/emoji/charts/full-emoji-list.html
# Always replace U+ with \U and Add enough leading zeros so that there are 8 hex digits total after \U.

unicode = u"\U0001F606\U0001F923"
print(unicode)

emoji = "\U00002764"

face_emoji = "\U0001F970"

ej = '\U0001F60E'

tongue_emoji = '\U0001F92A'

print(emoji)
print(face_emoji)

print(ej)

print(tongue_emoji)
```

```python
mess = 'Python is fun !\U0001f600'

print(mess)
```

```python
# Raw code 
raw_str = r"Bangladesh \n Me"

print(raw_str)
```

### Binary literals (0b1001)


```python
# boolean literals

val = True + 4     # True = 1

val2 = False + 10   # False = 0

print(val, type (val))

print(val2, type(val2))
```

```python
print(False + 1)
```

### octal literals (0o310)

```python
oct_val = 0o310
print(oct_val)
```

### hex-decimal (0x12c)

```python
hex_val = 0x12c 

print(hex_val, type(hex_val))
```
