### Assignment 1: Temperature Check

- Problem:
    - Write a program that takes the current temperature as input and prints:
    - ● "It's too cold!" if below 10°C
    - ● "It's cool outside" if between 10°C and 25°C
    - ● "It's hot outside!" if above 25°C

```python
# Assignment - 1: Temperature Check

current_temp = float(input("Enter the Current Temperature (°C): "))

if current_temp < 10:
    print("It's too cold!")
    
elif (current_temp >= 10) and (current_temp <= 25):
    print("It's cool outside")
    
else:
    print("It's hot outside!")
    
```

### Assignment 2: ATM Withdrawal System
- Problem:
    - Ask the user for their account balance and the amount they want to withdraw.
    
    - ● If the withdrawal amount is greater than the balance → print "Insufficient Balance"
    
    - ● If the withdrawal amount is less than or equal to balance → print "Transaction Successful" and show the remaining balance.

```python
# Assignment 2: ATM Withdrawal System

user_account_balance = float(input("Enter your Account Balance: "))

withdraw_balance = float(input("Enter your Withdraw amount: "))

if (withdraw_balance > user_account_balance):
    print("Insufficient Balance")
    
elif (withdraw_balance <= user_account_balance):
    print("Transaction Successful")
    
    rem_bal = user_account_balance - withdraw_balance
    
    print("Remaining Balance: ", rem_bal)

else:
    pass
```

### Assignment 4: Bus Ticket Discount

- Problem:
    - Ask the user for age and print ticket price:
    - ● Below 5 → Free
    - ● 5–18 → 50% Discount
    - ● 60 or above → 30% Discount
    - ● Others → Full Price

```python
# Assignment 4: Bus Ticket Discount
ticket_price = 500

user_age = int(input("Enter your age: "))

if (user_age <= 0) or (user_age >= 150):
    print("Invalid Age No Ticket.")
    
elif (user_age < 5):
    print("Bus Ticket is Free")
    
# 50% => 0.5
elif(user_age >= 5) and (user_age <= 18):
    discount = ticket_price *0.5
    print("Bus Ticket Discount: ", discount)

elif(user_age >= 60):
    discount = ticket_price *0.3
    print("Bus Ticket Discount: ", discount)
       
else:
    print("Full Price: ", ticket_price)
    
```

## Assignment 5: Restaurant Bill with Discount
- Problem:
    - Take total bill amount from the user:
    
    - ● If the bill is above 1000 → apply 10% discount

    - ● Otherwise → no discount Then print the final bill amount.


```python
# Assignment 5: Restaurant Bill with Discount

total_bill_amount = int(input("Enter Your Total Bill: "))

if total_bill_amount > 1000:
    discount = total_bill_amount * 0.10
    print("Discount amount: ", discount)
    
    total_bill_amount -= discount
    print("Final Bill amount: ", total_bill_amount)
    
else:
    print("Final Bill amount: ",total_bill_amount)

```

### Assignment 8: Leap Year Checker
- Problem:
    - Take a year as input and check:
    - ● If it is divisible by 400 → Leap Year
    - ● Else if divisible by 4 but not by 100 → Leap Year
    - ● Otherwise → Not a Leap Year

```python
# Assignment 8: Leap Year Checker

year = int(input("Enter a Year: "))

if (year % 400 == 0):
    print("Leap Year")
    
elif (year % 4 == 0) and (year % 100 != 0):
    print("Leap Year")
    
else:
    print("Not a Leap Year")
```

- Problem:

    - ● Above 200 units → 10 BDT per unit

```python
# Assignment 10: Electricity Bill Calculator

units = int(input("Enter Number of Units are Comsumed: "))

bill = 0

if units <= 100:
    bill = units * 5
    print("Total Payable amount: ", bill)
    
elif (units >= 101 ) and (units <= 200):
    bill = units * 7
    print("Total Payable amount: ", bill)
    
else:
    bill = units * 10
    print("Total Payable amount: ", bill)
```
