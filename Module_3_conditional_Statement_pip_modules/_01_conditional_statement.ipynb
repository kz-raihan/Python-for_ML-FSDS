### 2. Syntax of Conditional Statement

```python
'''  
for single condition  (if-else):

- if condition:
    if body execute when condition is True
    
- else:
    else body execute when if condition is False

'''
```

```python
""" 

# we want to build logic function

    - Actual Credentials
    Email: kzraihan@gmail.com
    pass : #kz123


"""
```

```python
# Example of if-else 

# - Actual Credentials

actual_mail = "kzraihan@gmail.com"
actual_password = "#kz123"


user_mail = input("Enter your email: ")
user_password = input("Enter your password: ")


if (user_mail == actual_mail) and (user_password == actual_password):
    print("Successfully login")
    
else:
    print("Please Try again")
```

#### Syntax of if-elif-else condition

```python
'''  
for multiple condition (if-elif-else):

- if condition:
    if body execute when condition is True

- elif condition:
    execute elif body when if conditon is False
    
- else:
    execute else body when if and elif condition is False
'''
```

```python
# Example : if-elif-else

# - Actual Credentials
actual_mail = "kzraihan@gmail.com"
actual_password = "#kz123"

# user enter
user_mail = input("Enter your email: ")
user_password = input("Enter your password: ")


if (user_mail == actual_mail) and (user_password == actual_password):
    print("Successfully login")
    
elif (user_mail != actual_mail) and (user_password == actual_password):
    print("Invalid Email") 


elif (user_mail == actual_mail) and (user_password != actual_password):
    print("InCorrect Password") 

else:
    print("Please Try again")
```

#### Nested condition

- condition (if-elif-else) inside another condition(if-elif-else) 

```python
'''  
for multiple condition (if-elif-else):

- if condition:
    if body execute when condition is True

- elif condition:
    execute elif body when if conditon is False
    
    if condition:
        --- execute ---
    elif condition:
        --- execute ----
    else:
        -- execute---
    
- else:
    execute else body when if and elif condition is False
'''
```

```python
# Example of Nested conditoin

# - Actual Credentials
actual_mail = "kzraihan@gmail.com"
actual_password = "#kz123"

# user enter
user_mail = input("Enter your email: ")
user_password = input("Enter your password: ")


# condition
if (user_mail == actual_mail) and (user_password == actual_password):
    print("Successfully login")
   
# nested condition 
elif (user_mail != actual_mail) and (user_password == actual_password):
    print("Invalid email..")
    user_mail = input("Enter your email Again: ")
    
    if user_mail == actual_mail:
        print("Successfully login")
    else:
        print("Invalid Email")
        

# Nested condition
elif (user_mail == actual_mail) and (user_password != actual_password):
    print("Incorrect Password..")
    user_password = input("Enter your password again: ")
    
    if user_password == actual_password:
        print("Successfully login")
    else:
        print("InCorrect Password")
     

else:
    print("Please Try again")
```

```python
# Exercise-1
# Find the minimum number from 3 given numbers

num1 = int(input("Enter 1st number: "))
num2 = int(input("Enter 2nd number: "))
num3 = int(input("Enter 3rd number: "))


if (num1 == num2 == num3):
    print("Numbers are Equal.")
    
elif (num1 <= num2) and (num1 <= num3):
    print("Number 1 is Minimum - ", num1)
    
elif (num2 <= num1) and (num2 <= num3):
    print("Number 2 is Mimimum - ", num2)
    
else:
    print("Number 3 is Minimum - ", num3)

```
