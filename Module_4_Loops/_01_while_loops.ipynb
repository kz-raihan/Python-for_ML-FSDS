- 3. Comapare/Match the number.

    - if guess > generate number

- 4. if guess == generate number

```python
# solution

# step_01: Generates a random number
import random

generate_num = random.randint(1,100)
generate_num
```

```python
# Step_02: Take a Input form user __Guess the number

guess_num = int(input("Enter your guess number: "))

# Step_3. Comapare/Match the number.

attem = 0
while guess_num != generate_num:
    
    if guess_num > generate_num:
        print("Wrong! Guess Lower")
        
    else:
        print("Wrong! Guess Higger")   
        
    
    guess_num = int(input("Enter your guess number: "))
    attem += 1
    
else:
    print("Correct Guess")
    print("Your Attem: ", attem)
```
