### 2. for loop with range()

- syntax of range()

- range(lower_bound : upper_bound : step)

- lower_bound = 0 [inclusive]
- upper_bound = 1 [Exclusive]

- step/jump parameter/ skip = 1 [By defalut]


```python
list(range(1,10,1))
```

```python
# for loop with range [in normal form]

for i in range(1,10):
    print(i, end = ' ')
```

```python
# for loop with range [step size positive]

for i in range(1, 11,2):
    print(i)
```

```python
# for loop with range [step size negative]

for i in range(1, 11,-1):
    print(i)
    
# what happen?
# - when step size is negative 
# then i start with upper bound and end with lower bound.
```

```python
# solutiont [when -1 negative step then exchange the upper and lower bound]

for i in range(10, 0, -1):
    print(i, end = " ")
```

```python
# for loop with step

for i in range(1, 30, 3):
    print(i,end =" ")
```

```python
# 7Ô∏è‚É£ Sum of numbers from 1 to 100

i = 1
sum = 0

while i <=100:
    sum += i
    i += 1

print("Summation: ", sum)
    
```

```python
# 8Ô∏è‚É£ Count how many even numbers are between 1 to 50

count_even = 0

i = 0

while i < 51:
    if i % 2 == 0:
        count_even += 1
    
    i += 1

print("Even Number: ", count_even)
```

```python
# 9Ô∏è‚É£ Take a user input N and print numbers from 1 to N

num = int(input("Enter Your number: "))

i = 1

while i <= num:
    print(i, end = " ")
    
    i += 1
```

```python
# üîü Print the multiplication table of 7

i = 1

while i <=10:
    print(f"{7} * {i} = {7 * i}")
    
    i += 1
```
