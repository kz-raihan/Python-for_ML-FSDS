```python
# (1) Email Validator Program

email = input("Enter your email: ")

if "@" in email:

    if email.endswith(".com") or email.endswith(".org"):
        print("Valid Email ")
    else:
        print("Invalid Email (must end with .com or .org)")
else:
    print("Invalid Email (missing '@')")
```

```python
# (2) Word Frequency Counter

review = input("Enter your review: ")

review = review.lower()


words = review.split()


word_count = {}


for word in words:
    if word in word_count:
        word_count[word] += 1
    else:
        word_count[word] = 1


print("Word Frequency:", word_count)
```

```python
# (3) Password Strength Checker

password = input("Enter your password: ")


has_upper = False
has_lower = False
has_digit = False
has_special = False

for ch in password:
    if ch.isupper():
        has_upper = True
    elif ch.islower():
        has_lower = True
    elif ch.isdigit():
        has_digit = True
    elif not ch.isalnum():
        has_special = True


if has_upper and has_lower and has_digit and has_special:
    print("Strong Password")
else:
    print("Weak Password (must include uppercase, lowercase, digit, and special character)")
```

```python
# (4) Username Generator using Loops

import random

first_name = input("Enter your first_name : ")

last_name = input("Enter your last_name : ")

username = ""

for i in range(3):
    if i < len(first_name):
        username += first_name[i].lower()


for i in range(-3, 0):
    if abs(i) <= len(last_name):
        username += last_name[i].lower()


random_num = random.randint(10, 99)
username += str(random_num)

print("Generated Username:", username)
```

```python
# (5) SMS Word Counter

message = input("Enter your message: ")


words = message.split()
word_count = len(words)


char_count = len(message)

if char_count > 160:
    status = "Exceeded Limit "
else:
    status = "Within Limit "

print("Words =", word_count)
print("Characters =", char_count)
print("Status =", status)
```

```python
#(6) Remove Duplicates from Contact List

contacts = input("Enter contact names (comma separated): ")


names = contacts.split(",")

clean_names = []
for name in names:
    clean_names.append(name.strip())


unique_names = []
for name in clean_names:
    if name not in unique_names:
        unique_names.append(name)


result = ", ".join(unique_names)

print("Clean Contact List:", result)
```

```python
# (7) Detect Suspicious Words

text = input("Enter your text: ").lower()

banned_words = ['spam', 'scam', 'fake']


for word in banned_words:
    if word in text:
        print("Warning: Suspicious content detected ")
        break
else:
    print("Text is clean ")
```

```python
# (10) Word Reversal for Data Encryption

sentence = input("Enter a sentence: ")


words = sentence.split()


reversed_words = []
for word in words:
    reversed_words.append(word[::-1])


reversed_sentence = " ".join(reversed_words)

print("Reversed Sentence:", reversed_sentence)
```
