### Loops Control Statement 

### OUTCOMES

- 1. break Control Statement

- 2. continue Control Statement

- 3. pass Control Statment

### 1. break control Statement

- The break statement exits the loops when condition is meet.

```python
# we only see first 5 number form 1-10

for i in range(1,11):
    if i == 6:
        break
    
    print(i)
```

```python
# break with list
# problem Statement: if list contain python then exit the loop

lst = ['Kz', 'Raihan', 'Banladesh', 'Python', 'AI']

for i in lst:
    if i == 'Python':
        break
    
    print(i, end=" ")
```

### 2. Continue Statments

- Skip (jump) the Current iteration and continues with the next.

- for loop has a continue statement (Default)

```python
# continue statement
# problem statement: if 5 present then continue (can't print)

for i in range(1,10):
    if i == 5:
        continue
    print(i, end= " ")
    
```

```python
# continue statement
# skip all event numbers form 1-100

for i in range(1,101):
    if i%2==0:
        continue
    
    print(i, end=" ")
```
