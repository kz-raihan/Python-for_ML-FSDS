### 1. Common Function

```python
# len() -> return the total number of character

promt = "My name is Kz"

print("Total length (character): ", len(promt))
```

```python
# max() -> returns the character with highest Unicode order.

# link of ascii unicode: https://www.ascii-code.com/characters/ascii-alphabet-characters

text = "Bangladesh is small and beautiful country"

print("The maximum character in string: ", max(text))
```

- why y?
    - Because (a = 97, and z = 122)
    - so, y is greater 

- note:
    - space ascii value: 32

    - how to check : ord("A")

```python
print("ASCII value of a:",ord('a'))
print("ASCII value of z:",ord('z'))
print("ASCII value of B:",ord('B'))
print("ASCII value of Space:",ord(' '))
```

```python
# min() -> Returns the character with the smallest unicode value in the string.

text = "Bangladesh is small and beautiful country"

print("Smaller Unicode character: ",min(text))
```

- the smaller unicode character for above text : space (32)

```python
# sorted()-> sorts every character in the string based on unicode(ASCII) oreder.

# - Return them as a list [by default -> ascending]

text = "Kz Raihan"

print("Sorted: ", sorted(text))
```

- for remember:
    - first sort Capital letters in ascending order
    - Then sort smaller letters in ascending order.


- NOTE:
    - sort by unicode order

```python
# sorted with reverse order
text = "Bangladesh is small country"
print("Reverse order: ", sorted(text, reverse = True))
```

### 2. Operations on Strings

```python
# Arithmetic Operations
#  - supported
#  - concatenation (+)
#  - Repetition  (*)


# concatenation(+)

first_name = "Kz"

last_name = "Raihan"

full_name = first_name + last_name

print("Full Name :", full_name)

# Repetition (*)

print("Hi! " * 3)

print("loading" + "." * 5)
```

```python
# Relational Operations

# compare two strings character-by-character based on unicode(ASCII)

print('Hi' >= 'hello')

print(ord('H'))  # H -> 72
print(ord('h'))  # h -> 104
```

```python
# Relational operation

print('Hello' != 'World')

print(ord('H'), ord('W'))
```

```python
# Relational operation
print('Kz' <= 'Raihan')

print(ord('K'), ord('R'))
```

```python
# Logical Operations (and, or, not)

print(" " and "Hi")
```

- Here, first string -> space
- 2nd string -> "Hi"
- and mean if two string are True then return last string which is "Hi"

```python
# or  operation

print('Kz' or 'raihan')
```

- if two string are True then return first string.

```python
# and , or

print("" and "Kz") # empty string-> off (0)

print("" or "Raihan") # or
```

- "and" means both(str) must be True.
    - if both str are True then return last str otherwise return False(empty string)

- "or" means at least one is True
    - if both True then return first string.
    - if at least one True then return True str. 

```python
# not logical operator

print(not('Hello'))

print(not(''))

# if True then return False and vice-versa.
```

```python
# Real__life Example

user_name = input("Enter Your name: ")

if not user_name:
    print("User name can not be empty")
    
else:
    print("Welcome", user_name)
```

```python
# Loops on Strings
name = "KzRaihan"

for i in name:
    print(i, end= ' ')
```

```python
# Real__life example: vowels count

sentence = "Bangladesh is beautiful"

vowel = 0

for char in sentence.lower():
    if char in 'aeiou':
        vowel += 1
        
print("Total Vowel: ", vowel)
```

```python
# Membership Operations
# - in 
# - not in
# check if something is exists inside a string (return True or False)


text = "Bangladesh is small"

print('b' in text)

print('d' not in text)

print('s' in text)
```

```python
# Real life__ example: check if email is valid

email = 'kz@gmail.com'

if "@" in email and "." in email:
    print("Email is valid")
    
else:
    print("Email is not valid")
```

```python
# Real_life example: Mask a phone number except the last 4 digit

phone = '01701234567'

masked = '*' * (len(phone)-4) + phone[-4:]

print(masked)
```
