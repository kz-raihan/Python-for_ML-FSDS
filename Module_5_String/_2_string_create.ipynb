### 2. How many ways to Creating String

- There are many ways to creating String

    - i. single quotation

    - ii. Double quotation

    - iii. Triple quotation

    - iv. build-in str() function

```python
# Example__1: Creating String with single quotation

country = 'Bagladesh is my Country'

print(country) 
```

```python
# problem statement__ single quotation

place = 'It's very beautiful'
place
```

```python
# Example__2: Double Quotation

place = "It's very beautiful"

print(place)
```

```python
# Problem_Statement__2: when we try to print multiline string 

germany_city = "My another Favorite country is Germany
- The Largest German Cities are
    - Berlin
    - Hamburg
"

print(germany_city)
```

```python
# Example__03: Triple Quotation

germany_city = """ My another Favorite country is Germany
- The Largest German Cities are
    - Berlin
    - Hamburg
    - Koln
    
"""

print(germany_city)
```
