## Topic: Virtual Enviroment Create

### Rules of Virtual Enviroment:

- Virtual env name  can't be same as package, libraries, funcion name.

- Create command
    - conda create -n enviroment_name python = verson of python -y

    - eg: 
        - conda create -n helmet python = 3.9 -y

        - where , helmet => env name
        - python verson = 3.9
        - -y => give the permission

    - where this (helmet) virtual enviroment create:
        - this pc -> local c -> conda vir(where anaconda is crete) [storage in c drive]


- 1. Create Virtual env:
    - conda create -n helmet python = 3.9 -y

- 2. Activate the venv name:
    - conda activate helmet

- 3. Deactivate the venv:
    - conda deactivate helmet (for outside the helmet env then command)

- 4. check the env list:
    - conda env list

- 5. Activate the base env:
    - conda activate base

- 6. To see the libraries inside env
    - pip list

- 7. install package inside env:
    - pip install pandas == 2.3.2

- 8. unstall package: 
    - pip unstall pandas

- 9. To Deleted env
    - first outside the env then comment
    - conda remove --name helmet --all

### Create global env on specific drive

- 1. Create on specific directory(go to specific directory and using command promt)
    
    -  conda create --prefix .\helmet python=3.9 -y 
    
    - where , helmet -> global enviroment

- 2. activate:
    - conda activate .\helmet

- NOTE:
    - The same way to others command
